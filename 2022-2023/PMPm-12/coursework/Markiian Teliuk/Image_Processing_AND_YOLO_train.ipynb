```python
#IMAGE_PROCESSING
import cv2
import os

directory = './RUSSIAN_PLANES_DB/train'



for filename in os.listdir(directory):
    if filename.endswith('.jpg') or filename.endswith('.jpeg') or filename.endswith('.png'):
        img = cv2.imread(os.path.join(directory, filename))
        
        #CANNY
        #edges=cv2.Canny(img,150,250)

        #LAPLACE
        #LESS_CONTRAST_EDGES (ver 1)
        denoise=cv2.GaussianBlur(img,(3,3),0)
        #MORE_CONTRAST_EDGES (ver 2)
        denoise=cv2.GaussianBlur(img,(1,1),0)
        
        #FURTHER_PROCESSING
        to_gray=cv2.cvtColor(denoise, cv2.COLOR_BGR2GRAY)
        res_lap=cv2.Laplacian(to_gray,cv2.CV_16S,ksize=3)

        cv2.imwrite(os.path.join('./RUSSIAN_PLANES_DB/train_lap_contrast/',filename), res_lap)
        #cv2.imwrite(os.path.join('./RUSSIAN_PLANES_DB/test_10_images/',filename), edges)
```
